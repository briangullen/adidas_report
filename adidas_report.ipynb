```python
import pandas as pd
import seaborn as sns
```

```python
adidas_df = pd.read_excel('Adidas US Sales Datasets.xlsx', skiprows=4)
```

```python
adidas_df = adidas_df.drop(columns='Unnamed: 0')
```

```python
sns.countplot(data=adidas_df, x='Region').set(title='Total Sales by Region')
```

```python
sns.relplot(data=adidas_df, x='Price per Unit', y='Operating Profit', hue='Sales Method').set(title='Operating Profit Compared with Pricer per Unit')
```

```python
sns.countplot(data=adidas_df, x='Sales Method', stat='percent').set(title='Percent of Sales by Method')
```

```python
ax = sns.barplot(data=adidas_df, x='Product', y='Units Sold')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40)
ax.set(title='Units Sold by Product')
```

```python
sns.boxplot(data=adidas_df, x='Retailer', y='Units Sold')
```
